### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



```python
tiendas = [tienda1, tienda2, tienda3, tienda4]
nom_tienda = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
sum_precios = [t["Precio"].sum() for t in tiendas]
tex_sum1 = ["" for t in tiendas]

for i in range(len(nom_tienda)):
  tex_sum1[i] = (f"{nom_tienda[i]}\n ($ {sum_precios[i]})")
  print(f"Total de Ventas {nom_tienda[i]}: $ {sum_precios[i]}")

print("")
plt.figure(figsize=(8,6))
plt.bar(tex_sum1, sum_precios)
plt.title("Facturación Ventas por Tienda")
plt.xlabel ("Tiendas")
plt.ylabel ("Facturación")

fig, ax = plt.subplots(figsize=(10,8))
ax.pie(sum_precios, autopct="%1.1f%%")
ax.set_title("Porcentaje de Ventas por Tienda")
ax.legend(nom_tienda)
plt.show()


```

# 2. Ventas por categoría

```python
cant_tienda = [t["Categoría del Producto"].value_counts().sort_values(ascending=False) for t in tiendas]
val_tienda = [t.groupby("Categoría del Producto")["Precio"].sum().sort_values(ascending=False) for t in tiendas]
tex_sum2 = [""]*8
tex_sum3 = [""]*8
for i in range(len(nom_tienda)):

  for j in range(len(tex_sum2)):
    tex_sum2[j] = (f"{cant_tienda[i].index[j]}\n ({cant_tienda[i][j]})\n")
    tex_sum3[j] = (f"{val_tienda[i].index[j]}\n ($ {val_tienda[i][j]})\n")

  print(f"Unidades Vendidas por Categoría {nom_tienda[i]}\n", cant_tienda[i].to_string().replace("Categoría del Producto",""),"\n")
  print(f"Venta Total por Categoria {nom_tienda[i]}\n",val_tienda[i].to_string().replace("Categoría del Producto",""),"\n")

  plt.figure(figsize=(12,6))
  plt.bar(tex_sum2, cant_tienda[i].values.tolist())
  plt.title("Unidades Vendidas por Categoria")
  plt.xlabel ("Categoria")
  plt.xticks(rotation=45, ha='right')
  plt.ylabel ("Ventas (Unidades)")

  plt.figure(figsize=(12,6))
  plt.bar(tex_sum3, val_tienda[i].values.tolist())
  plt.title("Monto Vendido por Categoria")
  plt.xlabel ("Categoria")
  plt.xticks(rotation=45, ha='right')
  plt.ylabel ("Ventas (Monto)")

  plt.show()
  print("")
```

# 3. Calificación promedio de la tienda


```python
calif = [t["Calificación"].mean().round(2) for t in tiendas]
tex_sum4 = ["" for t in tiendas]
for i in range(len(nom_tienda)):
  tex_sum4[i] = (f"{nom_tienda[i]}\n ({calif[i]}"")")
  print(f"Calificación Promedio de {nom_tienda[i]}: {calif[i]}")

print("")
plt.figure(figsize=(8,6))
plt.bar(tex_sum4, calif)
plt.title("Calificación Promedio por Tienda")
plt.xlabel ("Tiendas")
plt.ylabel ("Calificación")
plt.ylim(0,5)
plt.show()
```

# 4. Productos más y menos vendidos

```python
mas_vendidos = [t["Producto"].value_counts().head(3) for t in tiendas]
menos_vendidos = [t["Producto"].value_counts().tail(3) for t in tiendas]
tex_sum4 = [""]*3
tex_sum5 = [""]*3

for i in range(len(nom_tienda)):

  for j in range(len(tex_sum4)):
    tex_sum4[j] = (f"{mas_vendidos[i].index[j]}\n ({mas_vendidos[i][j]})\n")
    tex_sum5[j] = (f"{menos_vendidos[i].index[j]}\n ({menos_vendidos[i][j]})\n")

  print(f"Productos mas Vendidos {nom_tienda[i]}\n", mas_vendidos[i].to_string().replace("Producto",""),"\n")
  print(f"Productos menos Vendidos {nom_tienda[i]}\n", menos_vendidos[i].to_string().replace("Producto",""),"\n")

  plt.figure(figsize=(8,6))
  plt.bar(tex_sum4, mas_vendidos[i].values.tolist())
  plt.bar(tex_sum5, menos_vendidos[i].values.tolist(), color='red')
  plt.title("Productos Mas y Menos Vendidos")
  plt.xlabel ("Articulos")
  plt.xticks(rotation=45, ha='right')
  plt.ylabel ("Unidades Vendidas")
  plt.show()
  print("")
```

# 5. Envío promedio por tienda

```python
envio = [t["Costo de envío"].mean().round() for t in tiendas]
for i in range(len(nom_tienda)):
  tex_sum7[i] = (f"{nom_tienda[i]}\n ($ {envio[i]})")
  print(f"Costo de Envío Promedio {nom_tienda[i]}: $ {envio[i]}")

print("")
plt.figure(figsize=(8,6))
plt.bar(tex_sum7, envio)
plt.title("Costo de Envío Promedio por Tienda")
plt.xlabel ("Tiendas")
plt.ylabel ("Costo Envío")
plt.show()
```
